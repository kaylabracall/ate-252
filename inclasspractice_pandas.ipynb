```python
# Exploring PA Crash Data using pandas 

import pandas as pd
import numpy as np

crashes = pd.read_csv('crashtest.csv')
# crashes # pandas gave us an index. Index has to be immutable 

counties = crashes['County Name']
counties.dtype
print(counties.value_counts())

injuries = crashes['Injury Count']
print(injuries.describe())

fatalities = crashes['Fatality Count']
print(fatalities.value_counts())

rural_or_urban = crashes['Urban / Rural']
print(rural_or_urban.value_counts())

vehicles = crashes['Vehicle Count']
print(vehicles.max())

road_typesdf = crashes[['Interstate', 'State Road', 'Turnpike']]
print(road_typesdf)
```
